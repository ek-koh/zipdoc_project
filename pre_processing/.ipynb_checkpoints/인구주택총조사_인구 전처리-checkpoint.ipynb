# 1차 전처리
## 컬럼명 변경

```python
pop = pop.rename({'행정구역코드(시군구)': '행정구역코드_시군구'}, axis = 'columns')
```

```python
pop.dtypes
```

## 행정구역코드를 활용한 주소처리_1차

### '행정구역코드_시군구' 데이터타입 변환 : int -> object

```python
pop['행정구역코드_시군구'] = pop['행정구역코드_시군구'].apply(str)
```

### '행정구역코드_시군구' 앞 두자리 추출하여 '도시코드' 생성

```python
def first_two_str(row):
    return row['행정구역코드_시군구'][0:2]

pop['도시코드'] = pop.apply(first_two_str, axis=1)
```

```python
pop.head()
```

### '행정구역코드_시군구' 앞 네자리 추출하여 시군코드 생성

```python
def first_four_str(row):
    return row['행정구역코드_시군구'][0:4]

pop['시군코드'] = pop.apply(first_four_str, axis=1)
```

```python
pop.head()
```

### '행정구역코드_시군구', '도시코드', '시군코드' 데이터타입 변환 : object -> int

```python
pop['행정구역코드_시군구'] = pop['행정구역코드_시군구'].apply(int)
```

```python
pop['도시코드'] = pop['도시코드'].apply(int)
```

```python
pop['시군코드'] = pop['시군코드'].apply(int)
```

```python
pop.dtypes
```

```python
print(1110916-pop.count())
```

```python
pop.shape
```

### '1년전거주지행정구역코드'에서 9999로 입력된 값 null값으로 대체

```python
from numpy import NaN, NAN, nan
```

```python
pop.loc[pop['1년전거주지행정구역코드']==9999, '1년전거주지행정구역코드'] = nan
```

```python
pop[pop['1년전거주지행정구역코드']==9999]
```

```python
print(1110916-pop.count())
```

```python
pop.to_csv('../data/전처리완료/인구주택총조사_인구.csv', header=True, index=False)
```

```python
pop.shape
```

## 가구원번호 중복값 처리

```python
import pandas as pd
```

```python
pop = pd.read_csv('../data/전처리완료/인구주택총조사_인구_최종.csv')
```

```python
pop.info()
```

```python
len(pop['가구원번호'].unique())
```

```python
len(pop['가구번호'].unique())
```

```python
pop[pop.duplicated(['가구원번호'], keep=False)].sort_values(by=['가구원번호'], axis=0)
```

```python
pop2 = pop.drop_duplicates(['가구원번호'], keep='first')
```

- 키 값으로 사용할 수 있으나, 1208*2건의 중복 데이터가 발견되어 1208건을 제거함
- 인구주택_가구와 달리 데이터를 남긴 이유는 모든 컬럼이 같은 완벽히 중복된 데이터였기 때문
