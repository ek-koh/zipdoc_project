# 2017, 2018년 데이터 합치기

```python
import pandas as pd
```

## 데이터 로드 및 형태 확인

```python
data_2017 = pd.read_csv('../data/집닥_분석용 DB/9-1. 가계동향조사(지출부문)_가계동향조사(지출부문)(제공)_2017.csv')
```

```python
data_2018 = pd.read_csv('../data/집닥_분석용 DB/9-2. 가계동향조사(지출부문)_가계동향조사(지출부문)(제공)_2018.csv')
```

```python
data_2017.head()
```

```python
data_2018.head()
```

## 합쳐진 데이터프레임: concat

```python
concat = pd.concat([data_2017, data_2018], ignore_index=True)
```

# 데이터 형태 확인하기

## 노인가구, 모자가구, 맞벌이가구, 일반가구 중복 확인

```python
def sum_values(row):
    return row['노인가구']+row['모자가구']+row['맞벌이가구']+row['일반가구']
concat['sum_value'] = concat.apply(sum_values, axis=1)
```

```python
concat.head()
```

```python
concat.tail()
```

```python
concat['sum_value'].unique()
```

```python
concat.dtypes
```

```python
concat['sum_value']=concat['sum_value'].apply(int)
```

```python
concat.head()
```
