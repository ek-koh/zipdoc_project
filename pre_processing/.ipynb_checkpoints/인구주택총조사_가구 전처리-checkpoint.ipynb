## 행정구역코드를 활용한 주소처리_2차
### 행정구역코드 데이터와 merge

```python
code = pd.read_excel('../data/인구주택총조사_패널데이터/행정구역코드_2018.xlsx')
```

```python
code.head()
```

```python
apart2 = apart.merge(code, left_on='행정구역코드_시군구', right_on='코드')
```

```python
apart2.head()
```

```python
del apart2['코드']
apart2 = apart2.rename({'시도':'지역', '시군구': '주소_시군구'}, axis = 'columns')
```

```python
apart2.head()
```

```python
apart2['전체주소'] = apart2['지역']+' '+apart2['주소_시군구']
```

```python
code2 = pd.read_excel('../data/인구주택총조사_패널데이터/행정구역코드_2018_시군.xlsx')
```

```python
code2.head()
```

```python
apart3 = apart2.merge(code2, left_on='시군코드', right_on='코드')
```

## 가구번호 중복값 처리

```python
import pandas as pd
```

```python
gagu = pd.read_csv('../data/전처리완료/인구주택총조사_가구_최종.csv')
```

```python
gagu.info()
```

```python
len(gagu['가구번호'].unique())
```

```python
gagu[gagu.duplicated(['가구번호'], keep=False)].sort_values(by=['가구번호'], axis=0)
```

```python
gagu2 = gagu.drop_duplicates(['가구번호'],keep=False)
```

키 값으로 사용할 수 있으나, 262*2건의 중복값이 존재하여 전체 524건을 제거함.

전체를 제거한 이유는 인구주택_인구 데이터랑 merge해야 하는데 중복된 데이터가 완전히 중복이 아니라 가구번호는 중복, 이외의 데이터는 완전히 다른 데이터였기 때문

```python
gagu2.shape
```

```python
gagu2[gagu2.duplicated(['가구번호'], keep=False)].sort_values(by=['가구번호'], axis=0)
```

```python
len(gagu2['가구번호'].unique())
```

```python
gagu2.to_csv('../data/전처리완료/인구주택총조사_가구_final.csv', header=True, index=False)
```
